```python
import random
import matplotlib.pyplot as plt

#generation of 20 random integers
sequence_len = 20
min_val = 0
max_val = 100

random_num = [random.randint(min_val, max_val) for _ in range(sequence_len)]
print(f"generated numbers: {random_num}")

#deepest lake func.

def deepest_lake(arr):
    n = len(arr)
    max_depth = 0
    best_l = best_r = best_bottom = None

    for i in range(1, n-1):

        if arr[i] < arr[i-1]:

            left = i-1
            bottom = i

            while bottom < n-1 and arr[bottom+1] <= arr[bottom]:
                bottom += 1

            right = bottom
            while right < n-1 and arr[right+1] >= arr[right]:
                right += 1

            depth = min(arr[left], arr[right]) - arr[bottom]

            if depth > max_depth:
                max_depth = depth
                best_l = left
                best_r = right
                best_bottom = bottom

    return max_depth, best_l, best_bottom, best_r


depth, l, b, r = deepest_lake(random_num)
print("\ndeepest lake depth:", depth)
print("left boundary index:", l)
print("bottom index:", b)
print("right boundary index:", r)

#graph

x_val = list(range(sequence_len))
plt.plot(x_val, random_num, marker=' ', linestyle='-')

plt.xlabel("range of numbers")
plt.ylabel("generated numbers in range of (0-100)")

# підсвітка озера
if l is not None:
    lake_x = x_val[l:r+1]
    lake_y = random_num[l:r+1]

    plt.plot(lake_x, lake_y, linewidth=3, label="Deepest lake")
    plt.scatter([b], [random_num[b]], s=50, label="Lake bottom")


plt.title("Mountains and Deepest Lake")
plt.legend()
plt.grid()

plt.show()
```
