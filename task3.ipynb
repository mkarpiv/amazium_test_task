```python
import numpy as np

coin = np.array([0.8, 0.9, 0.1, 0.2, 0.3])

prob = np.ones(5) / 5
obs = ['H','T','H','H','H','T','T','H','H']
results = []

for r in obs:
    if r == 'H':
        l = coin
    else:
        l = 1 - coin

    # bayes'
    prob = prob * l
    prob = prob / prob.sum()

    future_H = np.sum(prob * coin)
    results.append(round(future_H, 2))

print(results)
```
